```python
import os

import os
from huggingface_hub import HfApi, create_repo
from pathlib import Path

# Configuration - Replace with your actual values
HF_USERNAME = "n1k1t427"  # Replace with your Hugging Face username
HF_TOKEN = "hf_token"   # Replace with your Hugging Face token

# Initialize Hugging Face API with token
api = HfApi(token=HF_TOKEN)

# Path to trained models directory
models_dir = Path("./trained_models")

# Check if the directory exists
if not models_dir.exists():
    print(f"Directory {models_dir} does not exist!")
else:
    # Iterate through all folders in trained_models
    for folder_name in os.listdir(models_dir):
        folder_path = models_dir / folder_name
        
        # Skip if not a directory
        if not folder_path.is_dir():
            continue
            
        print(f"Processing model: {folder_name}")
        
        try:
            # Create repository on Hugging Face Hub
            repo_id = f"{HF_USERNAME}/{folder_name}"
            
            # Create the repository (will skip if already exists)
            create_repo(repo_id=repo_id, token=HF_TOKEN, exist_ok=True)
            print(f"Created/verified repository: {repo_id}")
            
            # Upload all files in the folder
            api.upload_folder(
                folder_path=str(folder_path),
                repo_id=repo_id,
                repo_type="model"
            )
            print(f"Successfully uploaded {folder_name} to {repo_id}")
            
        except Exception as e:
            print(f"Error uploading {folder_name}: {str(e)}")
            
    print("Upload process completed!")
```
